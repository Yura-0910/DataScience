```kotlin
%use lets-plot

@file:DependsOn("com.opencsv:opencsv:5.7.1")
@file:DependsOn("org.jetbrains.lets-plot:lets-plot-kotlin-jvm:4.10.0")
//@file:DependsOn("org.jetbrains.lets-plot:lets-plot-kotlin-jvm:4.4.1")
//@file:DependsOn("org.jetbrains.lets-plot:lets-plot-common:4.5.0")
//@file:DependsOn("org.jetbrains.lets-plot:lets-plot-batik:4.5.0")

import com.opencsv.CSVReader
import java.io.FileReader
import java.time.LocalDateTime
import java.time.format.DateTimeFormatter
import org.jetbrains.letsPlot.*
import org.jetbrains.letsPlot.geom.*
import org.jetbrains.letsPlot.label.*
import org.jetbrains.letsPlot.themes.elementText
import org.jetbrains.letsPlot.themes.theme

val pathToFile: String = "/home/source/srcCode/DSAGitHub004/src/main/java/ru/lainer/DataScience/pressure.csv"
val fileReader = FileReader(pathToFile)
val csvReader = CSVReader(fileReader)
val resultOfReadCSV = csvReader.readAll()
csvReader.close()

val xAxis = mutableListOf<String>() // Список дат
val yAxis = mutableListOf<Int>()    // Список значений давления
for(current in 1 until resultOfReadCSV.size) {
    xAxis.add(resultOfReadCSV[current][0])
    yAxis.add(resultOfReadCSV[current][1].toInt()) //1-Верхнее давление, 2-Нижнее давление
}

// Создаём DataFrame-подобную структуру для lets-plot
val plotData = mapOf(
    "date" to xAxis,
    "pressure" to yAxis
)

// Строим график
val plot = letsPlot(plotData) +
        geomLine(color = "blue", size = 1.0) { x = "date"; y = "pressure" } +
        geomPoint(color = "red", size = 3.0) { x = "date"; y = "pressure" } +
        ggtitle("Динамика верхнего давления") +
        xlab("Дата и время") +
        ylab("Верхнее давление (мм рт. ст.)") +
        theme(axisTextX = elementText(angle = 45, hjust = 1))  // Наклон подписей дат

plot
```

```kotlin
// Можно добавить расчет корреляции

@file:DependsOn("com.opencsv:opencsv:5.7.1")
@file:DependsOn("org.nield:kotlin-statistics:1.2.1") //от 2018 года

import com.opencsv.CSVReader
import java.io.FileReader
import org.nield.kotlinstatistics.descriptiveStatistics
import org.nield.kotlinstatistics.median
import org.nield.kotlinstatistics.mode

val pathToFile: String = "/home/source/srcCode/DSAGitHub004/src/main/java/ru/lainer/DataScience/pressure.csv"
val fileReader = FileReader(pathToFile)
val csvReader = CSVReader(fileReader)
val resultOfReadCSV = csvReader.readAll()
csvReader.close()

val upperPressure = mutableListOf<Double>() // Список значений верхнего давления
val lowerPressure = mutableListOf<Double>() // Список значений нижнего давления

for(current in 1 until resultOfReadCSV.size) {
    upperPressure.add(resultOfReadCSV[current][1].toDouble()) //1-Верхнее давление
    lowerPressure.add(resultOfReadCSV[current][2].toDouble()) //2-Нижнее давление
}

//Среднее значение
val upperPressureAvg = upperPressure.average()
val lowerPressureAvg = lowerPressure.average()
println("Среднее значение верхнего давления = $upperPressureAvg")
println("Среднее значение нижнего давления = $lowerPressureAvg")

//Статистика
val statsUpperPressure = upperPressure.descriptiveStatistics
println("Верхнее давление (минимум):: ${statsUpperPressure.min}")
println("Верхнее давление (максимум):: ${statsUpperPressure.max}")
println("Верхнее давление (стандартнтое отклонение):: ${statsUpperPressure.standardDeviation}")
println("Верхнее давление (Колличество измерений давления):: ${statsUpperPressure.size}")
println("Верхнее давление (среднее значение):: ${statsUpperPressure.mean}")
println("Верхнее давление (медиана):: ${upperPressure.median()}")
val modesUpperPressure = upperPressure.mode()  // Возвращает Set (может быть несколько мод)
println("Наиболее частое верхнее давление (мода):: ${modesUpperPressure.joinToString(" и ")}")
```
